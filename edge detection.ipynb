Import modules

```python
import cv2
import matplotlib.pyplot as plt
```

```python
#to display images in the Jupyter Notebook

def display_image(img, title="Image", cmap_type=None):
    plt.imshow(img, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')  # Hides the axis values
    plt.show()
```

Read the image and resize: 

```python
img = cv2.imread('test.jpg')
img = cv2.resize(img, (300, 300))
```

## Display of the Original Image

```python
# convert from BGR to RGB for proper color display with matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

display_image(img_rgb, "Original Image")
```

## Display of the Gray Scale Image

```python
# convert to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# display the grayscale image
display_image(img_gray, "Grayscale Image", cmap_type='gray')
```

## Blur Image Display

```python
# blur the image for better detection
blur_img = cv2.GaussianBlur(img_gray, (3, 3), 0)

# display the blur image
display_image(blur_img, "Blurred Image", cmap_type='gray')
```

## Canny Edge Detection

```python
# canny edge detection
canny = cv2.Canny(blur_img, 100, 200)

# display the canny edge-detected image
display_image(canny, "Canny Edge Detection", cmap_type='gray')
```

## Laplacian Edge Detection

```python
# laplacian edge detection
laplacian = cv2.Laplacian(blur_img, cv2.CV_64F)

# display the laplacian edge-detected image
display_image(laplacian, "Laplacian Edge Detection", cmap_type='gray')
```
